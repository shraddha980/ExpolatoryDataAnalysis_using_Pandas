```python
#aggregate method : humko different columns pe different stats lagwane mai help karta hai jaise by the use of passing dictionary
genres.agg({
        'Runtime' : ['min','mean','max'],
        'IMDB_Rating' : 'mean',
        'No_of_Votes' : 'sum',
        'Gross' : 'sum',
        'Metascore' : ['min','mean','max']})
#Toh ye appko flexibility deta hai ki konse column k upar konsa function laga sakte hai
```

```python
#Find the highest rated movie of each genre
#we can solve this question using groupby
df = pd.DataFrame(columns=movies.columns)
```

```python
for group,data in genres:
    df = pd.concat([df, data[data['IMDB_Rating'] == data['IMDB_Rating'].max()]])
#concatenating the final data in this group
```

```python
def rank_movies(group):
    group['Rank'] = group['IMDB_Rating'].rank(ascending=False)
    return group
```

```python
rank_movies(movies)
```

```python
#Find normalized group rating
```

```python
def normalized_rating(group):
    group['Normalized_Rating'] = group['IMDB_Rating'] - group['IMDB_Rating'].min()/group['IMDB_Rating'].max() - group['IMDB_Rating'].min()
    return group
```

```python
genres.apply(normalized_rating)
#split, apply, groupby
```
